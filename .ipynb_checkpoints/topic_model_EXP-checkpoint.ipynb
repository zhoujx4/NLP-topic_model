# LDA

```python
import gensim
```

```python
def get_stopwords_list():
    """加载停用词表
    """
    stopwords = [line.strip() for line in open('data/hit_stopwords.txt',encoding='UTF-8').readlines()]
    return stopwords

def jieba_cut_remove_stopwords(text, stopwords):
    """结巴分词，并去除停用词
    """
    words = jieba.lcut(text)
    results = []
    for word in words:
        if word not in stopwords:
            results.append(word)
    return results

stopwords = get_stopwords_list()
df["content_cut"] = df["content"].apply(lambda x: jieba_cut_remove_stopwords(x, stopwords))
```

```python
documents = df["content_cut"].tolist()
dictionary = gensim.corpora.Dictionary(documents)
corpus = [dictionary.doc2bow(document) for document in documents]
```

```python
lda = gensim.models.LdaModel(num_topics=20,
                             corpus=corpus,
                             id2word=dictionary,
                             alpha="auto",
                             iterations=50,
                             random_state=4)
```

```python
perwordbound = lda.log_perplexity(corpus)# 好像这个函数的返回值，就是叫perwordbound
perplexity = pow(2,-perwordbound)# 这里好像也不是perplexity，只能作为一定的参考
```

```python
coherence_model_lda = gensim.models.CoherenceModel(model=lda, 
                                                   texts=documents, 
                                                   dictionary=dictionary, 
                                                   coherence='c_npmi')# 注意texts的取值不是corpus
```

```python
coherence_model_lda.get_coherence()
```

```python
import pyLDAvis
import pyLDAvis.gensim
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim.prepare(lda, corpus, dictionary)
```

```python
df = pd.read_csv("data/train.csv")
# df["content_cut"] = df["content"].apply(lambda x: jieba.lcut(x))
documents = df["content"].tolist()
```

```python
def tokenize_by_jieba(text):
    return jieba.lcut(text)
```

```python
from top2vec import Top2Vec

model = Top2Vec(documents=documents, 
                speed="learn", 
                workers=8, 
                tokenizer=tokenize_by_jieba)
```

```python
topic_sizes, topic_nums = model.get_topic_sizes()
```
